# Boston Dataset 

### Dataset loading

```python
#variant
v=13
```

```python
#http://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets
from sklearn import datasets
r = random.randint(1,20)
print(r)
data, target, coef = datasets.make_regression(n_samples = v*10, n_features=v,n_informative =round(v/5),
                                             noise = r,coef = True, random_state = 2)
```

### Feature plots

```python
pyplot.figure(figsize(20, 20))

plot_number = 0

for plot_number in range(v):
    pyplot.subplot(4, 4, plot_number + 1)
    pyplot.scatter(list(map(lambda x : x[plot_number], data)), target, color = 'b')
    pyplot.title('feature ' + str(plot_number))
    pyplot.xlabel(str(plot_number))
    pyplot.ylabel('target')
    plot_number += 1
```

### Linear regression model learning

```python
from sklearn import linear_model

linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(data, target)
predictions = linear_regressor.predict(data)
```

### Linear regression function

##### Learned Model

```python
s=""
for i in range(len(linear_regressor.coef_)):
    s+=str(round(linear_regressor.coef_[i],3))+"*x"+str(i)+" + "
s+=str(round(linear_regressor.intercept_,3))+"\n"
print(s)
```

##### Predefined

```python
s=""
for i in range(len(coef)):
    s+=str(round(coef[i],3))+"*x"+str(i)+" + "
s=s[:len(s)-2]
print(s)
```

### Basic Estimators

```python
from sklearn import metrics

mea = metrics.mean_absolute_error(target, predictions)
print ('MAE = ', mea)
msa = metrics.mean_squared_error(target, predictions)
print ('MSE = ', msa)
```

### Error plotting

```python
pyplot.figure(figsize(8, 8))
pyplot.scatter(predictions, target, color = 'b')
pyplot.xlabel('predictions')
pyplot.ylabel('target')
```
